```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
caminho_arquivo = "ocorrencias_sac.csv"

df = pd.read_csv(caminho_arquivo)
```

```python
# Limpar nomes das colunas
df.columns = df.columns.str.strip()
```

```python
# Converter a coluna de datas para o tipo datetime
df['data_ocorrencia'] = pd.to_datetime(df['data_ocorrencia'])
```

```python
df
```

Calcular probabilidade

```python
# Obter o número de ocorrências por dia
ocorrencias_por_dia = df['data_ocorrencia'].value_counts().sort_index()
```

```python
# Contar quantos dias tiveram 1, 2, 3... ocorrências
frequencia_ocorrencias = ocorrencias_por_dia.value_counts().sort_index()
```

```python
# Calcular a probabilidade de cada número de chamadas por dia
total_dias = ocorrencias_por_dia.count()
probabilidades = frequencia_ocorrencias / total_dias
```

```python
# Resultado em dataframe
distribuicao = pd.DataFrame({
    'num_chamadas': frequencia_ocorrencias.index,
    'frequencia_dias': frequencia_ocorrencias.values,
    'probabilidade': probabilidades.values
})
distribuicao
```

Criar histograma

```python
# Construir histograma com intervalos ("bins") definidos
# Usar numero de ocorrências calculado anteriormente
```

```python
# Definir os intervalos (bins) para o histograma
# Neste caso como todos os valores são 1, criamos uma faixa que engloba possíveis variações futuras
bins = [0, 1, 2, 3, 4, 5]
```

```python
# Criar o histograma com base nesses bins
plt.figure(figsize=(10, 6))
plt.hist(ocorrencias_por_dia, bins=bins, edgecolor='black', align='left', rwidth=0.8)

# Configurações do gráfico
plt.title('Histograma de Frequência de Ocorrências Diárias (SAC)')
plt.xlabel('Número de Chamadas por Dia (Faixas)')
plt.ylabel('Número de Dias')
plt.xticks(bins)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
